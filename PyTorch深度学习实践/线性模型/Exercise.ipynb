```python
# Import necessary library to draw the graph
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Prepare the train set
# y = 2x + 1
x_data = np.array([1.0, 2.0, 3.0])
y_data = np.array([3.0, 5.0, 7.0])
```

```python
# Define the model
def forward(x):
    return x * w + b
```

```python
# Define the loss function
def loss(x, y):
    y_pred = forward(x)
    return (y_pred - y) * (y_pred - y)
```

```python
# Generate parameter range
W = np.arange(0.0, 4.1, 0.1)
B = np.arange(0.0, 4.1, 0.1)
w, b = np.meshgrid(W, B)

# List mse_list save the cost values of each w
mse_list = []
```

```python
# Compute cost value
# For each sample in train set, the loss function values were computed
loss_sum = 0
for x_val, y_val in zip(x_data, y_data):
    y_pred_val = forward(x_val)
    loss_val = loss(x_val, y_val)
    loss_sum += loss_val
```

```python
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(w, b, loss_sum / float(x_data.size), cmap='viridis')

ax.set_xlabel('w')
ax.set_ylabel('b')
ax.set_zlabel('Loss')
ax = Axes3D(fig)

plt.show()
```
